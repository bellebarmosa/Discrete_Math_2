Consider the following experiment:
1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

---

## Question 1: 
Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$

$\textbf{Answers}$

To prove that the expected value $ E[y] = \sum y \cdot P(y) = n \cdot p_i $, we start with the definition of expected value:

$ E[y] = \sum y \cdot P(y) $

Given the joint probability distribution $ P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} $, where $ y_i $ represents the number of trials for which the outcome falls into cell $ i $, and $ P_i $ represents the probability that the outcome falls into cell $ i $, we have:

$ E[y] = \sum y \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} $

Now, let's rewrite the expected value in terms of the individual probabilities $ p_i $:

$ E[y] = \sum (y_1 \cdot P_1 + y_2 \cdot P_2 + y_3 \cdot P_3 + \dotsb + y_n \cdot P_n) \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} $

$ = \sum \left( y_1 \cdot \frac{n!}{y_1!} \cdot P_1^{y_1} \cdot P_2^{y_2} \cdot P_3^{y_3} \dotsb P_n^{y_n} + \dotsb + y_n \cdot \frac{n!}{y_n!} \cdot P_1^{y_1} \cdot P_2^{y_2} \cdot P_3^{y_3} \dotsb P_n^{y_n} \right) $

$ = \sum \left( n \cdot P_1^{y_1} \cdot P_2^{y_2} \cdot P_3^{y_3} \dotsb P_n^{y_n} \right) $

$ = n \cdot \left( P_1^1 + P_2^2 + P_3^3 + \dotsb + P_n^n \right) $

$ = n \cdot (p_1 + p_2 + p_3 + \dotsb + p_n) $

Since $ p_1 + p_2 + p_3 + \dotsb + p_n = 1 $ (as stated in the problem), we have:

$ E[y] = n \cdot 1 = n \cdot p_i $

This proves that the expected value $ E[y] = \sum y \cdot P(y) = n \cdot p_i $.

## Question 2: Given the following data

\begin{gather}
\begin{bmatrix}
\begin{array}{c|c}
Age & Proportion \\
18-24 & 0.18 \\
25-34 & 0.23 \\
35-44 & 0.16 \\
45-64 & 0.27 \\
65-100 & 0.16 \\
\end{array}
\end{bmatrix}
\end{gather}

If 500 adults are sampled  randomly, find the probability rhat the sample contains 100 person between 18 and 24, 200 between, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64.. What is the expected value for to obtain a person in the 65 and above?

$\textbf{Answers}$

To find the probability that the sample contains 100 persons between the ages of 18 and 24, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64, we'll use the multinomial probability formula.

Given:
- $n = 500$ (total number of adults sampled)
- $k = 5$ (number of age groups)
- $p_1 = 0.18$, $p_2 = 0.23$, $p_3 = 0.16$, $p_4 = 0.27$, $p_5 = 0.16$ (proportions for each age group)

The probability mass function (PMF) for a multinomial distribution is:

$P(y) = \frac{n!}{y_1!y_2!y_3!y_4!y_5!} p_1^{y_1} p_2^{y_2} p_3^{y_3} p_4^{y_4} p_5^{y_5}$

Where:
- $y_1$, $y_2$, $y_3$, $y_4$, $y_5$ are the counts for each age group

For the given scenario, we have:
- $y_1 = 100$ (for ages 18-24)
- $y_2 = 200$ (for ages 25-34)
- $y_3 = 0$ (no persons between ages 35-44)
- $y_4 = 200$ (for ages 45-64)
- $y_5 = 500 - (y_1 + y_2 + y_3 + y_4) = 500 - (100 + 200 + 0 + 200) = 0$ (for ages 65-100)

Substitute these values into the PMF and calculate:

$P(y) = \frac{500!}{100!200!0!200!0!} (0.18)^{100} (0.23)^{200} (0.16)^{0} (0.27)^{200} (0.16)^{0}$

$P(y) = \frac{500!}{100!200!200!} (0.18)^{100} (0.23)^{200} (0.27)^{200}$

The expected value to obtain a person in the 65 and above age group is given by the sum of the products of each outcome and its probability:

$E[Y_5] = n \times p_5 = 500 \times 0.16 = 80$

So, the expected number of persons aged 65 and above is 80.

$$
E[y] = \sum_{y = 0}^{n} y\binom{n}{y}p^{y}(1-p)^{n-y}
$$

By the binomial theorem

$$
(p + [1-p]) ^ {n} = \sum_{y = 0}^{n} \binom{n}{y}p^{y}(1-p)^{n-y} = 1
$$


Differentiate both sides by $p$
$$

Multiply both sides by $p$

$$

Therefore

$$

$$

P(Y \geq 9) = \binom{10}{9} p^9 (1 - p)^{10 - 9} + \binom{10}{10} p^{10} (1 - p)^{10 - 10}

$$
P(Y \geq 9) = \binom{10}{9} (0.30)^9 (0.70)^{10 - 9} + \binom{10}{10} (0.30)^{10} (0.70)^{10 - 10}

A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.

$\textbf{Answers}$

Let...

$Y$ = Number of seedlings in a square meter that follows a Poisson distribution of $\lambda = 5$

Let us use the probability mass function...

$$
P(Y = y) = \frac{e^{-\lambda} \lambda^y}{y!}
$$

$$
P(Y = y) = \frac{e^{-5} 5^y}{y!}
$$

$Y = 0$, denotes the probability that none of the regions will contain a seeling.

$$
P(Y = 0) = \frac{e^{-5} 5^0}{0!} = e^{-5}
$$

$$
P(Y = 0) = (e^{-5})^{10} = e^{-50}
$$

Therefore...

The probability that none of the regions will contain any seedlings is $\approx e^{-50}$
